```python
# Split the dataset into training and testing reviews
norm_train_reviews = train_df['review'][:30000]
norm_test_reviews = test_df['review'][10000:]

# Display the first review from the training set and a review at index 45005 in the test set
print("First review in training data:")
print(norm_train_reviews.iloc[0])

print("\nReview at index 5 in test data (45005 adjusted to fit the test set range):")
print(norm_test_reviews.iloc[5])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize the TfidfVectorizer
tv = TfidfVectorizer(min_df=0.0, max_df=1.0, binary=False, ngram_range=(1, 3))

# Transformed train reviews (processing the first 30,000 reviews)
tv_train_reviews = tv.fit_transform(norm_train_reviews[:30000])

# Transformed test reviews (using the fitted vectorizer)
tv_test_reviews = tv.transform(norm_test_reviews)

# Print the shapes of the transformed train and test data
print('Tfidf_train:', tv_train_reviews.shape)
print('Tfidf_test:', tv_test_reviews.shape)
```

```python
from sklearn.preprocessing import LabelBinarizer

# Initialize LabelBinarizer
lb = LabelBinarizer()

# Apply LabelBinarizer on train and test sentiment columns
sentimentdata_train = lb.fit_transform(train_df['sentiment'])
sentimentdata_test = lb.transform(test_df['sentiment'])  # Use transform on test data (to keep consistent with the training data)

# Print the shapes of the transformed sentiment data for both train and test sets
print('Sentiment data (train) shape:', sentimentdata_train.shape)
print('Sentiment data (test) shape:', sentimentdata_test.shape)
```

```python
# Split sentiment data based on your specified range
train_sentiment = sentimentdata_train[:30000]  # First 30,000 reviews for training
test_sentiment = sentimentdata_test[10000:]  # Reviews from index 10,000 onward for testing

# Print the sentiment data for both train and test sets
print('Train Sentiment Data:', train_sentiment)
print('Test Sentiment Data:', test_sentiment)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# 1. Naive Bayes (MultinomialNB) model
nb = MultinomialNB()
nb_tfidf = nb.fit(tv_train_reviews, train_sentiment)
nb_tfidf_predict = nb.predict(tv_test_reviews)
nb_tfidf_accuracy = accuracy_score(test_sentiment, nb_tfidf_predict)
nb_tfidf_report = classification_report(test_sentiment, nb_tfidf_predict, target_names=['Positive', 'Negative'])

# Output Naive Bayes results
print("Naive Bayes Accuracy (TF-IDF):", nb_tfidf_accuracy)
print("Classification Report (Naive Bayes):\n", nb_tfidf_report)

# 2. Logistic Regression model
lr = LogisticRegression(penalty='l2', max_iter=1000, C=1, random_state=42)
lr_tfidf = lr.fit(tv_train_reviews, train_sentiment)
lr_tfidf_predict = lr.predict(tv_test_reviews)
lr_tfidf_accuracy = accuracy_score(test_sentiment, lr_tfidf_predict)
lr_tfidf_report = classification_report(test_sentiment, lr_tfidf_predict, target_names=['Positive', 'Negative'])

# Output Logistic Regression results
print("Logistic Regression Accuracy (TF-IDF):", lr_tfidf_accuracy)
print("Classification Report (Logistic Regression):\n", lr_tfidf_report)

# 3. k-NN (k-Nearest Neighbors) model
knn = KNeighborsClassifier()
knn_tfidf = knn.fit(tv_train_reviews, train_sentiment)
knn_tfidf_predict = knn.predict(tv_test_reviews)
knn_tfidf_accuracy = accuracy_score(test_sentiment, knn_tfidf_predict)
knn_tfidf_report = classification_report(test_sentiment, knn_tfidf_predict, target_names=['Positive', 'Negative'])

# Output k-NN results
print("k-NN Accuracy (TF-IDF):", knn_tfidf_accuracy)
print("Classification Report (k-NN):\n", knn_tfidf_report)

# 4. Random Forest model
rf = RandomForestClassifier(random_state=42)
rf_tfidf = rf.fit(tv_train_reviews, train_sentiment)
rf_tfidf_predict = rf.predict(tv_test_reviews)
rf_tfidf_accuracy = accuracy_score(test_sentiment, rf_tfidf_predict)
rf_tfidf_report = classification_report(test_sentiment, rf_tfidf_predict, target_names=['Positive', 'Negative'])

# Output Random Forest results
print("Random Forest Accuracy (TF-IDF):", rf_tfidf_accuracy)
print("Classification Report (Random Forest):\n", rf_tfidf_report)
```

```python
# Create a DataFrame to display accuracy in tabular format
accuracy_data = {
    'Model': ['Naive Bayes', 'Logistic Regression', 'k-NN', 'Random Forest'],
    'Accuracy': [nb_tfidf_accuracy, lr_tfidf_accuracy, knn_tfidf_accuracy, rf_tfidf_accuracy]
}

accuracy_df = pd.DataFrame(accuracy_data)

# Print the accuracy table
print(accuracy_df)
```

**CONCLUSION**

Naive Bayes and Logistic Regression perform very similarly across all three variations, with Logistic Regression slightly outperforming Naive Bayes in the second and third variations.
k-NN sees a performance increase with the third variation (sublinear scaling and bigrams), achieving the highest accuracy of 0.7945 compared to the other variations.
Random Forest performs consistently, with minor improvements in the third variation (0.8549 from 0.8493 in the first).
The second variation, which incorporates stopwords removal, generally improves the accuracy for Logistic Regression, but it slightly reduces performance for k-NN compared to the first variation.
The third variation (with sublinear scaling and bigrams) benefits models like k-NN and Logistic Regression, where it achieves slightly better results overall. However, the difference is marginal for most models.

In summary, the third variation with sublinear scaling and bigrams is the most effective
